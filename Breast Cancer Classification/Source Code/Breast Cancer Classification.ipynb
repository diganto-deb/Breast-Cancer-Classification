# BREAST CANCER CLASSIFICATION

## Problem Statement

 > Predicting if the cancer diagnosis is being malignant or benign based on several observations/features
 
 > 30 features are used,examples :
 
     > radius (mean of distances from center to points on the perimeter)
     > texture (standard deviation of gray scale values)
     > perimeter
     > smoothness
     > compactness
     > concavity
     > concave points
     > symmetry
     > fractal dimension

> Datasets are linearli separabe using all 30 input features

> Number of Instances :569

> Class distribution : 212 Malignant, 357 Benign

>Target class :
        
        - Malignant
        - Benign

## Importing Data

## VISULIZING THE DATA

```python
sb.heatmap(df.isnull(),cmap='viridis')
plt.show()
```

## Model Training

```python
X=df.drop('target',axis=1)
y=df['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=5)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
```

```python
svc=SVC()
```

```python
svc.fit(X_train,y_train)
```

## Evaluating the model

```python
pred=svc.predict(X_test)
```

```python
cm=confusion_matrix(y_test,pred)
```

```python
sb.heatmap(cm,annot=True)
```

```python
min_train=X_train.min()
range_train=(X_train-min_train).max()
X_train_scaled=(X_train-min_train)/range_train
```

```python
#Without Scaling
sb.scatterplot(x=df['mean area'],y=df['mean smoothness'],hue=y_train)
plt.show()
```

```python
#With Scaling the data
sb.scatterplot(x=X_train_scaled['mean area'],y=df['mean smoothness'],hue=y_train)
plt.show()
```

## Improving the model 

```python
min_test=X_test.min()
range_test=(X_test-min_test).max()
X_test_scaled=(X_test-min_test)/range_test
```

```python
svc.fit(X_train_scaled,y_train)
```

```python
pred=svc.predict(X_test_scaled)
```

```python
cm=confusion_matrix(y_test,pred)
```

```python
sb.heatmap(cm,annot=True,cmap='viridis')
```

```python
print(classification_report(y_test,pred))
```

## Improving the model again 

```python
param_grid={'C':[0.1,1,10,100],'gamma':[1,0.1,0.01,0.001],'kernel':['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(SVC(),param_grid,refit=True,verbose=4)
```

```python
grid.fit(X_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
grid_pred=grid.predict(X_test_scaled)
```

```python
cm=confusion_matrix(y_test,grid_pred)
```

```python
sb.heatmap(cm,annot=True,cmap='viridis')
```

```python
print(classification_report(y_test,grid_pred))
```
